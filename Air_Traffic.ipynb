##Filtering Operation Airline

```python
count_values= df['Operating Airline'].value_counts() 
count_values_less_than_20 = count_values[count_values<=20]  
```

```python
def extraction(x):
  if x in count_values_less_than_20:
    return False
  else:
    return True
```

```python
df1=df[df['Operating Airline'].apply(extraction)]
```

```python
df1['Operating Airline'].value_counts()
```

##Filtering Operating Airline IATA Code

```python
count_values_IATA= df1['Operating Airline IATA Code'].value_counts() 
count_values_IATA_less_than_20 = count_values_IATA[count_values_IATA<=20]  
```

```python
def extraction2(x):
  if x in count_values_IATA_less_than_20:
    return False
  else:
    return True
```

```python
df2=df1[df1['Operating Airline IATA Code'].apply(extraction2)]
```

```python
df2['Operating Airline IATA Code'].value_counts()
```

##Filtering Published Airline


```python
count_values_Published_Airline= df2['Published Airline'].value_counts() 
count_values_Published_Airline_less_than_20 = count_values_Published_Airline[count_values_Published_Airline<=20]  
```

```python
def extraction3(x):
  if x in count_values_Published_Airline_less_than_20:
    return False
  else:
    return True
```

```python
df3=df2[df2['Published Airline'].apply(extraction3)]
```

```python
df3['Published Airline'].value_counts()
```

##Filtering Published Airline IATA Code

```python
count_values_Published_Airline_i= df3['Published Airline IATA Code'].value_counts() 
count_values_Published_Airline_i_less_than_20 = count_values_Published_Airline_i[count_values_Published_Airline_i<=20]  
```

```python
def extraction4(x):
  if x in count_values_Published_Airline_i_less_than_20:
    return False
  else:
    return True
```

```python
df4=df3[df3['Published Airline IATA Code'].apply(extraction4)]
```

```python
df4['Published Airline IATA Code'].value_counts()
```

```python
df4.isnull().sum()
```

```python
df4.shape
```

```python
df4['Month'].value_counts()
```

##Convert Year into Months

```python
df4['Year']=(current_time.year-df4['Year'])*12
```

```python
def month_converter(month):
    months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
    return months.index(month) + 1
```

```python
df4['Month'] = df4['Month'].apply(month_converter)
```

##Clculate Total Month

```python
df4['months']=df4['Year'] + df4['Month']
```

```python
df4
```

##Drop Null Value

```python
df4= df4.dropna(axis='rows')
```

```python
df4 = df4.drop(['Year','Month'],axis='columns')
```

```python
df4
```

```python
df4.isnull().sum()
```

##Convert Categorical Data Numeric

```python
d_Ap = pd.get_dummies(df4['Activity Period'])
d_Oa = pd.get_dummies(df4['Operating Airline'])
d_Opi = pd.get_dummies(df4['Operating Airline IATA Code'])
d_Pa = pd.get_dummies(df4['Published Airline'])
d_Pai = pd.get_dummies(df4['Published Airline IATA Code'])
d_Gs = pd.get_dummies(df4['GEO Summary'])
d_Gr = pd.get_dummies(df4['GEO Region'])
d_atc = pd.get_dummies(df4['Activity Type Code'])
d_pc = pd.get_dummies(df4['Price Category Code'])
d_tr = pd.get_dummies(df4['Terminal'])
d_ba = pd.get_dummies(df4['Boarding Area'])
d_aatc = pd.get_dummies(df4['Adjusted Activity Type Code'])
```

##Concatinate all the dummies with dataset and remove Unnecessary Columns

```python
df5 = pd.concat((df4,d_Ap,d_Oa,d_Opi,d_Pa,d_Pai,d_Gs,d_Gr,d_atc,d_pc,d_tr,d_ba,d_aatc),axis='columns')
```

```python
df6 = df5.drop(['Activity Period','Operating Airline','Operating Airline IATA Code','Published Airline','Published Airline IATA Code','GEO Summary','GEO Region','Activity Type Code','Price Category Code','Terminal','Boarding Area','Adjusted Activity Type Code'],axis='columns')
```

```python
df6.shape
```

```python
df6.head()
```

```python
df6.shape
```

```python
X=df6.drop('Adjusted Passenger Count',axis='columns')
Y=df6.iloc[:,1:2].values
```

```python
X
```

```python
Y
```

##Split Data

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y)
```

```python
xtrain.columns
```

##Apply Model

```python
from sklearn.linear_model import LinearRegression
lmodel = LinearRegression()
```

```python
lmodel.fit(xtrain,ytrain)
```

##Error

```python
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(ytrain,lmodel.predict(xtrain)))
print(mean_absolute_error(ytest,lmodel.predict(xtest)))
```

##Score

```python
print(lmodel.score(xtrain,ytrain))
print(lmodel.score(xtest,ytest))
```
